Рассмотрим сложную математическую функцию на отрезке [1, 15]:

f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)


```python
import numpy as np
import matplotlib.pyplot as plt
```

Сформируйте систему линейных уравнений (то есть задайте матрицу коэффициентов A и свободный вектор b) для многочлена первой степени, который должен совпадать с функцией f в точках 1 и 15. Решите данную систему с помощью функции scipy.linalg.solve. Нарисуйте функцию f и полученный многочлен.

```python
def f(x):
    return np.sin(x/5.0)*np.exp(x/10.0) + 5*np.exp(-x/2.0)
```

```python
f(1)
```

```python
f(15)
```

```python
def f_1(x,w) :
    return w[0] + x*w[1]
```

```python
def f_aproximation(f, x, w):
    return np.array([f(xx,w) for xx in x])
```

```python
x_1 = 1
x_2 = 15
a = np.array([[1,x_1],[1,x_2]])
b = np.array([f(x_1), f(x_2)])
w = np.linalg.solve(a,b)
```

```python
import matplotlib.pyplot as plt
x = np.arange(1,15, 0.01)
print f(x)
plt.plot(x, f(x), x, f_aproximation(f_1,x,w))
plt.xlabel(r'$x$') 
plt.ylabel(r'$f(x)$') 
plt.grid(True)
plt.show()
```

Повторите те же шаги для многочлена второй степени, который совпадает с функцией f в точках 1, 8 и 15.


```python
def f_2(x,w) :
    return w[0] + x*w[1] + x**2 * w[2]
```

```python
x_1 = 1
x_2 = 8
x_3 = 15
a = np.array([[1,x_1, x_1**2],[1, x_2, x_2**2],[1, x_3, x_3**2]])
b = np.array([f(x_1), f(x_2), f(x_3)])
w = np.linalg.solve(a,b)
```

```python
x = np.arange(1,15, 0.01)
plt.plot(x, f(x), x, f_aproximation(f_2,x,w))
plt.xlabel(r'$x$') 
plt.ylabel(r'$f(x)$') 
plt.grid(True)
plt.show()
```

Повторите те же шаги для многочлена третьей степени, который совпадает с функцией f в точках 1, 4, 10 и 15.


```python
def f_3(x,w) :
    return w[0] + x*w[1] + x**2 * w[2] + x**3 * w[3]
```

```python
x_1 = 1
x_2 = 4
x_3 = 10
x_4 = 15
a = np.array([[1,x_1, x_1**2, x_1**3],[1, x_2, x_2**2, x_2**3],[1, x_3, x_3**2, x_3**3], [1, x_4, x_4**2, x_4**3]])
b = np.array([f(x_1), f(x_2), f(x_3), f(x_4)])
w = np.linalg.solve(a,b)
```

```python
x = np.arange(1,15, 0.01)
plt.plot(x, f(x), x, f_aproximation(f_3,x,w))
plt.xlabel(r'$x$') 
plt.ylabel(r'$f(x)$') 
plt.grid(True)
plt.show()
```
